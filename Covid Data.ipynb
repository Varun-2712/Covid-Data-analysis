### Let's Import the modules 

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
print('Modules are imported.')
```

```python
### importing covid19 dataset
covid=pd.read_csv("covid19_Confirmed_dataset.csv")

covid
```

```python
#### shape
print(covid.shape)
```

```python
covid.drop(columns=["Lat","Long"],inplace=True)
```

```python
covid_data=covid.groupby("Country/Region").sum()
covid_data.head()
```

```python
covid_data.loc["India"]
covid_data.loc["India"].plot()
plt.legend()
```

```python
## Calculating the first derivative
covid_data.loc["India"].diff().plot()
covid_data.loc["India"].diff()
```

```python
country=list(covid_data.index)
max_inf=[]
for i in country:
    max_inf.append(covid_data.loc[i].diff().max())
```

```python
happ=pd.read_csv("worldwide_happiness_report.csv")
happ.set_index("Country or region",inplace=True)
happ.head()
```

```python
happ.shape
```

```python
final_data=happ.join(covid_data,how="inner")
print(final_data.shape)
```
